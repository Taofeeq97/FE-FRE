```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
X = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
#Applying PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python

model_orig_pca = LogisticRegression()
model_orig_pca.fit(X_train_pca, y_train)
y_pred_orig_pca = model_orig_pca.predict(X_test_pca)
accuracy_orig_pca = accuracy_score(y_test, y_pred_orig_pca)  # Correct!
print("Accuracy of original model:", accuracy_orig_pca)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
X_train_lda = lda.fit_transform(X_train_scaled, y_train)
X_test_lda = lda.transform(X_test_scaled)
```

```python
explained_variance = lda.explained_variance_ratio_
print(explained_variance)
```

```python
model_orig_lda = LogisticRegression()
model_orig_lda.fit(X_train_lda, y_train)
y_pred_orig_lda = model_orig_lda.predict(X_test_lda)
accuracy_orig_lda = accuracy_score(y_test, y_pred_orig_lda)
print("Accuracy of original model:", accuracy_orig_lda)
```

```python
#Applying KPCA for features reduction
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=10, kernel='rbf')
X_train_kpca = kpca.fit_transform(X_train_scaled)
X_test_kpca = kpca.transform(X_test_scaled)
```

```python
model_orig_lpkca = LogisticRegression()
model_orig_lpkca.fit(X_train_kpca, y_train)
y_pred_orig_lpkca = model_orig_lpkca.predict(X_test_kpca)
accuracy_orig_lpkca = accuracy_score(y_test, y_pred_orig_lpkca)
print("Accuracy of original model:", accuracy_orig_lpkca)
```

```python
#apply QDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
qda = QDA()
X_train_qda = qda.fit(X_train_scaled, y_train)
x_pred_qda = qda.predict(X_test_scaled)
accuracy_qda = accuracy_score(y_test, x_pred_qda)
print("Accuracy of QDA model:", accuracy_qda)
```
